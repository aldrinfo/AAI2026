## Execute ReACT Generation

````
def run_react_generation(task):    '''Execute ReACT code generation'''        print("\n" + "="*70)    print("ReACT CODE GENERATION")    print("="*70)    print(f"\nTASK: {task}\n")        # Get ReACT response    print("[Generating with ReACT...]")    response = client.messages.create(        model="claude-3-5-sonnet-20241022",        max_tokens=2000,        messages=[{"role": "user", "content": REACT_PROMPT.format(task=task)}]    )        react_output = response.content[0].text    print(react_output)        # Extract code    print("\n" + "-"*70)    code_match = re.search(r"```python\n(.*?)\n```", react_output, re.DOTALL)        if code_match:        code = code_match.group(1)        print("\n[Executing generated code...]")        try:            exec_globals = {}            exec(code, exec_globals)                        # Run tests if available            if 'test_solution' in exec_globals:                print("\n✓ Tests:")                exec_globals['test_solution']()            elif 'main' in exec_globals:                print("\n✓ Output:")                exec_globals['main']()                        return True        except Exception as e:            print(f"✗ Execution error: {e}")            return False    else:        print("✗ Could not extract code")        return Falseprint("✓ ReACT function ready")
````
