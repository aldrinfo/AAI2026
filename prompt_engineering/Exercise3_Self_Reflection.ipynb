# Exercise 3: Self-Reflection Prompt for Improving Output**Goal:** Ask AI to critique and improve its own summary using explicit criteria.

```
import anthropicclient = anthropic.Anthropic()INITIAL_SUMMARY_PROMPT = '''Summarize this technical document in 150-200 words.Focus on main concepts and key takeaways.DOCUMENT:{document}Provide a clear, structured summary.'''SELF_CRITIQUE_PROMPT = '''You are a critical editor. Evaluate this summary against explicit criteria:CRITERIA:1. Clarity: Is language clear and accessible?2. Completeness: Are all main concepts covered?3. Conciseness: Is it within 150-200 words without fluff?4. Accuracy: Is information accurate?5. Structure: Is it well-organized?6. Actionability: Can readers extract actionable insights?ORIGINAL SUMMARY:{summary}Provide detailed critique for each criterion.Identify 2-3 specific improvements.'''IMPROVE_PROMPT = '''Based on this critique, rewrite the summary to address gaps.ORIGINAL:{summary}CRITIQUE:{critique}Rewrite while maintaining 150-200 words.Output ONLY the improved summary, no explanations.'''print("✓ Prompts loaded")
```

## Execute Self-Reflection Cycle

```
def run_self_reflection(document):    '''Execute the improvement cycle'''        print("\n" + "="*70)    print("SELF-REFLECTION IMPROVEMENT CYCLE")    print("="*70)        # Step 1: Initial summary    print("\n[Step 1] Generating initial summary...")    resp1 = client.messages.create(        model="claude-3-5-sonnet-20241022",        max_tokens=300,        messages=[{"role": "user", "content": INITIAL_SUMMARY_PROMPT.format(document=document)}]    )    original = resp1.content[0].text.strip()    print(f"\n📝 ORIGINAL ({len(original.split())} words):\n{original}")        # Step 2: Self-critique    print("\n" + "-"*70)    print("\n[Step 2] Self-critique...")    resp2 = client.messages.create(        model="claude-3-5-sonnet-20241022",        max_tokens=500,        messages=[{"role": "user", "content": SELF_CRITIQUE_PROMPT.format(summary=original)}]    )    critique = resp2.content[0].text.strip()    print(f"\n🔍 CRITIQUE:\n{critique}")        # Step 3: Improved version    print("\n" + "-"*70)    print("\n[Step 3] Generating improved summary...")    resp3 = client.messages.create(        model="claude-3-5-sonnet-20241022",        max_tokens=300,        messages=[{"role": "user", "content": IMPROVE_PROMPT.format(summary=original, critique=critique)}]    )    improved = resp3.content[0].text.strip()    print(f"\n✨ IMPROVED ({len(improved.split())} words):\n{improved}")        print("\n" + "="*70)        return original, critique, improvedprint("✓ Reflection function ready")
```

## Test with MLOps Document

```
test_doc = '''Machine Learning Operations (MLOps) is an engineering discipline that combines Machine Learning (ML), DevOps, and data engineering. It aims to design, build, and manage reproducible, testable, and evolvable ML systems. MLOps focuses on automating the deployment, monitoring, and management of ML models in production environments. Key components include data pipeline orchestration, model training automation, continuous integration/continuous deployment (CI/CD) for models, monitoring and observability, and governance frameworks. MLOps enables teams to collaborate more effectively, reduces time-to-market for ML projects, improves model reliability, and ensures compliance with regulatory requirements. By applying software engineering best practices to ML systems, organizations can achieve better scalability, maintainability, and performance of their AI initiatives.'''original, critique, improved = run_self_reflection(test_doc)
```
